```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import pyplot
import pandas as pd
from shapely.geometry import Point
import geopandas as gpd
from geopandas import GeoDataFrame

```

```python
df1 = pd.read_pickle(r'C:\Users\maaza\miniconda3\pkl data\data_1km.pkl')
```

### TIR(Thermal Infrared) detectors and Fire Channels(S7 to S9 and F1, F2)  have a resolution of 1km at nadir.
###  BT=Brightness Temperature


```python
F1=df1[["F1_BT","time"]]
F2=df1[["F2_BT","time"]]
S7=df1[["S7_BT","time"]]
S8=df1[["S8_BT","time"]]
S9=df1[["S9_BT","time"]]

fig,axs=plt.subplots(2,3,figsize=(8,6))
#ax.plot(df1.time,df1.F1_BT)
axs[0, 0].plot(df1.time,df1.F1_BT)
axs[0, 0].set_title('F1')
axs[0, 1].plot(df1.time,df1.F2_BT, 'tab:orange')
axs[0, 1].set_title('F2')
axs[1, 0].plot(df1.time,df1.S7_BT, 'tab:green')
axs[1, 0].set_title('S7')
axs[1, 1].plot(df1.time,df1.S8_BT, 'tab:red')
axs[1, 1].set_title('S8')
axs[1, 2].plot(df1.time,df1.S9_BT, 'tab:purple')
axs[1, 2].set_title('S9')

#for ax in axs.flat:
 #   ax.set(xlabel='x-label', ylabel='y-label')
```

```python
lat_min=df1['lat'].min()
lat_max=df1['lat'].max()
lat_max=df1['lat'].max()
lon_min=df1['lon'].min()
lon_max=df1['lon'].max()
print(lat_min, lat_max, lon_min, lon_max)
```

```python

#geometry will take longitude as x and latitude as y and put it in Point(xy) because of the for loop in zip function we will get
#only one Point at a time and 
geometry = [Point(xy) for xy in zip(df1['lon'], df1['lat'])]
gdf = GeoDataFrame(df1, geometry=geometry)   

#this is a simple map that goes with geopandas
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
#cmap = plt.colormaps.get_cmap('RdYlGn')
username = input("F1_BT\nF2_BT\nS7_BT\nS8_BT\nS9_BT\nChoose one channel from above and write the name here for visualization:")
if username=='F1_BT':
    gdf.plot(ax=world.plot(figsize=(10, 6)),column=df1['F1_BT'], marker='o',cmap="hot_r", markersize=15,legend=True)
elif username=='F2_BT':
    gdf.plot(ax=world.plot(figsize=(10, 6)),column=df1['F2_BT'], marker='o',cmap="hot_r", markersize=15,legend=True);
elif username=='S7_BT':
    gdf.plot(ax=world.plot(figsize=(10, 6)),column=df1['S7_BT'], marker='o',cmap="hot_r", markersize=15,legend=True);
elif username=='S8_BT':
    gdf.plot(ax=world.plot(figsize=(10, 6)),column=df1['S8_BT'], marker='o',cmap="hot_r", markersize=15,legend=True);
elif username=='S9_BT':
    gdf.plot(ax=world.plot(figsize=(10, 6)),column=df1['S9_BT'], marker='o',cmap="hot_r", markersize=15,legend=True);
else:
    print("Not found")
```
